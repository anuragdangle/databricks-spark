```
from pyspark.sql.functions import udf,col
from pyspark.sql.types import StringType

def udfCase(case):
    if case>50:
        return 'high'
    else:
        return 'low'
    
# caseudf = udf(udfCase, StringType())
# or
caseudf = udf(lambda z : udfCase(z) , StringType())

caseDf.select(col('confirmed'),caseudf(col('confirmed'))).show()
# or
# caseDf.select(col('confirmed'),caseudf(col('confirmed').alias('highorlow'))).show()
```
